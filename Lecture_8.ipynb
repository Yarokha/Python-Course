### Contour Plot

```python
x = np.linspace(-2, 2, 201)
y = np.linspace(-2, 2, 201)
z = np.exp(-x**2-y**2)
```

```python
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2-Y**2)
cfp = plt.contourf(X, Y, Z)
```

```python
cfp = plt.contourf(X, Y, Z, cmap=plt.magma())
plt.colorbar(cfp)
```

```python
cfp = plt.contourf(X, Y, Z, cmap=plt.magma(), vmin=0.45)
plt.colorbar(cfp)
```

```python
cfp = plt.contourf(X, Y, Z, cmap=plt.magma(), vmin=0.45, vmax=0.7)
plt.colorbar(cfp)
```

```python
plt.figure(layout="constrained", figsize=(16,3))
plt.subplot(141)
cfp = plt.contourf(X, Y, Z,  cmap=plt.magma())
plt.colorbar(cfp)
plt.subplot(142)
cfp = plt.contourf(X, Y, Z, levels=3,  cmap=plt.magma())
plt.colorbar(cfp)
plt.subplot(143)
cfp = plt.contourf(X, Y, Z, levels=11,  cmap=plt.magma())
plt.colorbar(cfp)
plt.subplot(144)
cfp = plt.contourf(X, Y, Z, levels=101,  cmap=plt.magma())
plt.colorbar(cfp)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6) )

cfp1 = axes[0,0].contourf(X, Y, Z, levels=101, vmin=z.min(), vmax=z.max(), cmap=plt.magma())
cfp2 = axes[0,1].contourf(X, Y, Z * 0.7, levels=101, vmin=z.min(), vmax=z.max(), cmap=plt.magma())
cfp3 = axes[1,0].contourf(X, Y, Z * 0.5, levels=101, vmin=z.min(), vmax=z.max(), cmap=plt.magma())
cfp4 = axes[1,1].contourf(X, Y, Z * 1.2, levels=101, vmin=z.min(), vmax=z.max(), cmap=plt.magma())

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.12, 0.03, 0.75])
fig.colorbar(cfp1, cax=cbar_ax, ticks=np.linspace(0, 1, 11))

plt.show()
```

### 4D Plot

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import cm


T = Z * np.sin(x)* np.exp(y)
minn, maxx = T.min(), T.max()
norm = matplotlib.colors.Normalize(minn, maxx)
m = plt.cm.ScalarMappable(norm=norm, cmap='jet')
fcolors = m.to_rgba(T)


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.set_title("4D")
plot_surf = ax.plot_surface(X, Y, Z,  facecolors=fcolors)
ax.set(xlabel=(r'x'), ylabel=(r'y'), zlabel=(r'z'))
cb = plt.colorbar(m, ax=ax, location='right', cax = ax.inset_axes([1.15, 0., 0.05, 1]))
cb.set_label('t')
```

### Polar Plot

```python
r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2]) 
ax.set_rlabel_position(-22.5)
ax.grid(True)
plt.show()
```

### Inset Plot

```python
x = np.linspace(-10,10,100)
y1 = np.sin(x)*x
y2 = np.cos(x)

fig, ax = plt.subplots(dpi=100)
ax.plot(x, y1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("y = x sin(x)")

ax_in = ax.inset_axes([0.3, 0.7, 0.45, 0.2])
ax_in.plot(x, y2)
ax_in.set_xlabel("x")
ax_in.set_ylabel("y")
ax_in.set_title("y = cos(x)")
plt.show()
```
