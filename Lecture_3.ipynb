## Slicing

### Basic Slicing

```python
arr = np.arange(15)
print("Original array:", arr)

print("\nStep slicing examples:")
print(f"arr[::2] -> every 2nd element: {arr[::2]}")
print(f"arr[1::2] -> every 2nd element starting from index 1: {arr[1::2]}")
print(f"arr[::3] -> every 3rd element: {arr[::3]}")
print(f"arr[2:10:2] -> from index 2 to 9, step 2: {arr[2:10:2]}")
print(f"arr[::-1] -> reversed array: {arr[::-1]}")
print(f"arr[10:2:-2] -> from index 10 to 3, step -2: {arr[10:2:-2]}")
```

### Boolean Indexing

```python
arr = np.arange(15)
print("Original array:", arr)

print("\nBoolean indexing examples:")
print(f"arr > 5 -> boolean mask: {arr > 5}")
print(f"arr[arr > 5] -> elements greater than 5: {arr[arr > 5]}")
print(f"arr[(arr > 3) & (arr < 10)] -> elements between 3 and 10: {arr[(arr > 3) & (arr < 10)]}")
print(f"arr[arr % 2 == 0] -> even elements: {arr[arr % 2 == 0]}")
print(f"arr[arr % 3 == 0] -> elements divisible by 3: {arr[arr % 3 == 0]}")
```

### Fancy Indexing

```python
arr = np.arange(15)
print("Original array:", arr)

print("\nFancy indexing examples:")
indices = [1, 3, 5, 7, 9]
print(f"indices = {indices}")
print(f"arr[indices] -> elements at specific indices: {arr[indices]}")
print(f"arr[[0, 2, 4, 6, 8, 10, 12, 14]] -> even indices: {arr[[0, 2, 4, 6, 8, 10, 12, 14]]}")
print(f"Using negative indices [-1, -3, -5]: {arr[[-1, -3, -5]]}")
```

```python
# 2D fancy indexing
arr_2d = np.arange(20).reshape(4, 5)
print("Original 2D array:")
print(arr_2d)

print("\n2D fancy indexing:")
row_indices = [0, 2, 3]
col_indices = [1, 3, 4]
print(f"Row indices: {row_indices}")
print(f"Column indices: {col_indices}")
print(f"arr[row_indices, col_indices] -> specific elements: {arr_2d[row_indices, col_indices]}")
print("arr[row_indices] -> specific rows:")
print(arr_2d[row_indices])
print("arr[:, col_indices] -> specific columns:")
print(arr_2d[:, col_indices])
```

## Approximation

```python
x = np.arange(-10, 11, 1)
y = x**2 + 4 * x - 1
y_noisy = y + np.random.normal(0,10,x.shape[0])
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y, label='y(x)')
plt.plot(x, y_noisy, label='y(x) + noise')
plt.legend()
```

```python
deg_ =  2
poly_coef = np.polyfit(x, y_noisy, deg=deg_,)
y_fit = np.poly1d(poly_coef)
print(y_fit)
```

```python
plt.plot(x, y, label='y(x)')
plt.plot(x, y_noisy, label='y(x) + noise')
plt.plot(x, y_fit(x), label='y_fit(x)')
plt.legend()
```

# Linear Algebra

### Vector Operations

```python
a_vec = np.array([1, 2, 3])
b_vec = np.array([4, 5, 6])

dot_product = np.dot(a_vec, b_vec)
cross_product = np.cross(a_vec, b_vec)
outer_product = np.outer(a_vec, b_vec)
magnitude_a = np.linalg.norm(a_vec)
angle = np.arccos(dot_product / (np.linalg.norm(a_vec) * np.linalg.norm(b_vec)))

print("Vector a:", a_vec)
print("Vector b:", b_vec)
print("Dot product:", dot_product)
print("Cross product:", cross_product)
print("Outer product:\n", outer_product)
print("Magnitude of a:", round(magnitude_a, 2))
print(f"Angle between vectors: {round(angle, 2)} radians")
```

### Matrix Operations

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix multiplication:\n", A @ B)
print("Element-wise multiplication:\n", A * B)
print("A + B:\n", A + B)
print("Transpose of A:\n", A.T)
```

### Eigenvalues and Eigenvectors

```python
A_eigen = np.array([[4, -2], [1, 1]])
eigenvalues, eigenvectors = np.linalg.eig(A_eigen)

print("Matrix for analysis:\n", A_eigen)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:")
print(np.round(eigenvectors, 3))

v1 = eigenvectors[:, 0]
lambda1 = eigenvalues[0]
print("\nVerification (A*v = λ*v):")
print(" A*v1:", np.round(A_eigen @ v1, 3))
print(" λ1*v1:", np.round(lambda1 * v1, 3))


v2 = eigenvectors[:, 1]
lambda2 = eigenvalues[1]
print("\nVerification (A*v = λ*v):")
print(" A*v2:", np.round(A_eigen @ v2, 3))
print(" λ2v2:", np.round(lambda2 * v2, 3))
```

### Solving Linear Systems

Example:

$$
\begin{cases}
2x + y = 5\\
x + 3y = 6
\end{cases}
$$

```python
A_system = np.array([[2, 1], [1, 3]])
b_system = np.array([5, 6])

solution = np.linalg.solve(A_system, b_system)
print("Coefficient matrix:\n", A_system)
print("Right-hand side:", b_system)
print("Solution: x = {:.2f}, y = {:.2f}".format(solution[0], solution[1]))
print("Verification (A @ x) == b:", np.abs(A_system @ solution - b_system) < 1e-9)
```

### Random Number Generation

```python
np.random.seed(42)

uniform = np.random.uniform(0, 1, 10)
normal = np.random.normal(0, 1, 10)
binomial = np.random.binomial(10, 0.5, 10)
poisson = np.random.poisson(5, 10)

arr_to_shuffle = np.arange(10)
np.random.shuffle(arr_to_shuffle)

print("Random Distributions:")
print(f"   Uniform (0-1): {np.round(uniform, 3)}")
print(f"   Normal (μ=0, σ=1): {np.round(normal, 3)}")
print(f"   Binomial (n=10, p=0.5): {binomial}")
print(f"   Poisson (λ=5): {poisson}")
print("Shuffled array:", arr_to_shuffle)
```

### Statistical Functions

```python
np.random.seed(42)
data = np.random.normal(0, 1, 1000)

print("Generated 1000 samples from normal distribution")
print("Statistical Summary:")
print(f"   Mean: {np.mean(data):.3f}")
print(f"   Median: {np.median(data):.3f}")
print(f"   Std Dev: {np.std(data):.3f}")
print(f"   Variance: {np.var(data):.3f}")
print(f"   Min: {np.min(data):.3f}")
print(f"   Max: {np.max(data):.3f}")
print(f"   Range: {np.ptp(data):.3f}")
print("Percentiles:")
percentiles = np.percentile(data, [25, 50, 75])
print(f"   25th: {percentiles[0]:.3f}")
print(f"   50th: {percentiles[1]:.3f}")
print(f"   75th: {percentiles[2]:.3f}")
```

### Correlation and Covariance

```python
np.random.seed(42)
data = np.random.multivariate_normal(
    mean=[0, 0, 0],
    cov=[[1, 0.8, 0.7],
         [0.8, 1, 0.6],
         [0.7, 0.6, 1]],
    size=100
)

correlation = np.corrcoef(data.T)
covariance = np.cov(data.T)

print("Dataset with 3 correlated variables (100 samples)")
print("Correlation matrix:")
print(np.round(correlation, 3))
print("Covariance matrix:")
print(np.round(covariance, 3))
```
