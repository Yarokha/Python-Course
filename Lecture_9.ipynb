## ODE

```python
f = Function('f')
x = Symbol('x')
ode1 = Eq(f(x).diff(x, 2) + f(x), 0)
solution1 = dsolve(ode1, f(x))
```

```python
ode1
```

```python
solution1
```

```python
ode2 = Eq(f(x).diff(x) - f(x), 0)
solution2 = dsolve(ode2, f(x), ics={f(0): 1})
```

```python
ode2
```

```python
solution2
```

```python
diffeq = Eq(f(x).diff(x, x) - 2*f(x).diff(x) + f(x), sin(x))
solve_dif = dsolve(diffeq, f(x))
```

```python
diffeq
```

```python
solve_dif
```
